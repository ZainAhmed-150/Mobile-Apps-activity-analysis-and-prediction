```python
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')
x=df['age']
y=df['Total_Duration_mins']
plt.scatter(x,y,color='c')
plt.title('Age v/s Total Duration')
plt.xlabel('Age')
plt.ylabel('Total Duration')
plt.show()
```

### Maximum population belongs to the age range 0-10 and Total Duration belongs to the range of 0-200 mins

```python

from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')
x1=df['age']
y1=df['Average_Duration_mins']

plt.scatter(x1,y1,color='g')
plt.title('Age v/s Average Duration')
plt.xlabel('Age')
plt.ylabel('Average Duration')
plt.show()

```

### Maximum population belongs to the age range 0-10 and Average Duration belongs to the range of 0-5 mins

```python
import matplotlib.pyplot as plt
import numpy as np

y_stack = np.row_stack((df['Average_Duration_mins'],df['Total_Duration_mins'])) 
builds=df['age']
fig = plt.figure(figsize=(11,8))
ax1 = fig.add_subplot(111)

ax1.plot(builds, y_stack[0,:], label='Component 1', color='c', marker='o')
ax1.plot(builds, y_stack[1,:], label='Component 2', color='g', marker='o')


handles, labels = ax1.get_legend_handles_labels()
lgd = ax1.legend(handles, labels, loc='upper center', bbox_to_anchor=(1.15,1))
ax1.grid('on')
plt.show()
plt.savefig('smooth_plot.png')
```
